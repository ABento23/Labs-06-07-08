```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

df=pd.read_excel('https://github.com/masterfloss/data/blob/main/exerciseInt.xlsx?raw=true')
df=df.dropna()
```

```python
y = df['Income']
X = df.drop(['country','Income'],axis=1)
```

```python
model = sm.OLS(y,X).fit()
model.summary()
```

```python
#3
# The variables IQ, BMI female and BMI male are not statistically significant, since their
# p-value is greater than 0.1.
# There is statistical evidence to consider the variables Education expenditures per capita
# and quality of Life significant.
```

```python
y2 = df['Income']
X2 = df[['IQ','Education expenditures per capita']]

model = sm.OLS(y2,X2).fit()
model.summary()
```
